```python
df['Ship Mode'].value_counts()
```

```python
x=df['Ship Mode'].value_counts().index
y=df['Ship Mode'].value_counts().values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,8))
plt.pie(y, labels=x, startangle=40, autopct='%0.2f%%')
plt.legend(loc=1)
plt.title('Ship Mode')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Ship Mode',data=df,hue='Category')
plt.title('Distribution of Shipping Modes by Category')
plt.show()
```

```python
sns.countplot(x='Segment',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.countplot(x='Category',data=df, hue='Sub-Category')
plt.title('Distribution of Sub-Categories with in Each Category') 
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Order Year'] = df['Order Date'].dt.year
```

```python
df.info()
```

```python
df['Order Year'].value_counts()
```

```python
sns.countplot(x='Order Year',data=df)
plt.show()
```

```python
sns.barplot(x='Category',y='Profit',data=df,estimator='sum')
plt.show()
```

```python
df['State'].value_counts()[:5]
```

```python
top_5_states = df['State'].value_counts().nlargest(5)
```

```python
top_5_states
```

```python
plt.figure(figsize=(8, 9))
plt.pie(top_5_states, labels=top_5_states.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 States by States')
plt.axis('equal') 
plt.show()
```
